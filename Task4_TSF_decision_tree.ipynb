## Problem statement and Solution:
    
For the given ‘Iris’ dataset, create the Decision Tree classifier and
visualize it graphically. The purpose is if we feed any new data to this
classifier, it would be able to predict the right class accordingly.
Dataset :
In the folder 


```python
# Importing Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
dataset=pd.read_csv('iris.csv')
dataset.head()
```

```python
print(dataset['Species'].unique())
print('------------------||-----------------------')
print(dataset.info())
print('------------------||-----------------------')
print(dataset.isnull().sum())
```

```python
dataset.describe()
```

```python
x=dataset[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
y=dataset['Species'].values
```

```python
print(x[:5,:5])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print('Accuracy is',accuracy_score(pred,y_test))
print('\nClassification: \n ',classification_report(pred,y_test))
```

```python
from sklearn import tree
feature=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
plt.figure(figsize=(40,30))
tree.plot_tree(dt.fit(x,y),feature_names=feature,filled=True,precision=3,proportion=True,rounded=True)
plt.show()
```
